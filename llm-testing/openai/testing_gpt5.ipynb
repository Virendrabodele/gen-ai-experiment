###Testing Generated Code

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create sample DataFrame
data = [
    {'Department': 'Engineering', 'Employee': 'Alice',  'Salary': 120000},
    {'Department': 'Engineering', 'Employee': 'Bob',    'Salary': 110000},
    {'Department': 'Engineering', 'Employee': 'Charlie','Salary': 125000},
    {'Department': 'Sales',       'Employee': 'David',  'Salary': 90000},
    {'Department': 'Sales',       'Employee': 'Eva',    'Salary': 95000},
    {'Department': 'HR',          'Employee': 'Frank',  'Salary': 70000},
    {'Department': 'HR',          'Employee': 'Grace',  'Salary': 72000},
    {'Department': 'Marketing',   'Employee': 'Heidi',  'Salary': 85000},
    {'Department': 'Marketing',   'Employee': 'Ivan',   'Salary': 88000},
    {'Department': 'Finance',     'Employee': 'Judy',   'Salary': 98000},
    {'Department': 'Finance',     'Employee': 'Karl',   'Salary': 102000},
]

df = pd.DataFrame(data)

# 2. Calculate average salary per department
avg_salary = df.groupby('Department')['Salary'].mean().sort_values(ascending=False)

print("Average salary by department:")
print(avg_salary.round(2))

# 3. Plot bar chart of average salaries per department
plt.figure(figsize=(8, 5))
avg_salary.plot(kind='bar', color='#4C72B0')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

###  3D Earth Visualization with Plotly

```python
# 1. Setup LLM via Openai
gpt5 = ChatOpenAI(
    model="gpt-5",
    openai_api_key=api_key
)

# Prompt for generating a 3D Earth visualization
prompt = """
Write a Python script using the `plotly.graph_objects` and `pandas` libraries to create an interactive 3D visualization of the Earth.

The script should perform the following steps:
1.  Create a 3D sphere that visually represents the Earth. A blue sphere with a specific texture or color scale is ideal.
2.  Define the latitude and longitude for a few major cities: London, New York, Tokyo, and Sydney.
3.  Convert these latitude and longitude coordinates into 3D Cartesian (x, y, z) coordinates to be plotted on the sphere's surface.
4.  Plot the cities as distinct markers on the 3D globe.
5.  The final output should be an interactive 3D plot that can be rotated and zoomed.
"""

# Let's get the model to generate the script
response = gpt5.invoke(prompt)

# The generated code will be printed below.
print("--- Generated 3D Earth Visualization Code ---")
print(response.content)
print("\n--- End of Generated Code ---")
print("\nIMPORTANT: Copy the code generated above into a new Colab cell and run it to see the interactive visualization.")
```

###Testing Code

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go


def latlon_to_xyz(lat_deg, lon_deg, radius=1.0):
    lat = np.radians(lat_deg)
    lon = np.radians(lon_deg)
    x = radius * np.cos(lat) * np.cos(lon)
    y = radius * np.cos(lat) * np.sin(lon)
    z = radius * np.sin(lat)
    return x, y, z


def build_earth_sphere(radius=1.0, n=100):
    # Parametric sphere (u: longitude, v: latitude)
    u = np.linspace(0, 2 * np.pi, n)
    v = np.linspace(-np.pi / 2, np.pi / 2, n)
    u, v = np.meshgrid(u, v)

    x = radius * np.cos(v) * np.cos(u)
    y = radius * np.cos(v) * np.sin(u)
    z = radius * np.sin(v)

    # Use z as surfacecolor to create a subtle latitudinal gradient
    surface = go.Surface(
        x=x,
        y=y,
        z=z,
        surfacecolor=z,
        colorscale="Blues",
        cmin=-radius,
        cmax=radius,
        showscale=False,
        opacity=0.95,
        lighting=dict(ambient=0.5, diffuse=0.8, fresnel=0.2, specular=0.5, roughness=0.2),
        lightposition=dict(x=100, y=200, z=0),
    )
    return surface


def build_city_markers(df_cities, radius=1.0, offset=0.01):
    # Slightly offset markers above the surface so they’re clearly visible
    r_markers = radius * (1.0 + offset)
    x, y, z = latlon_to_xyz(df_cities["lat"].values, df_cities["lon"].values, r_markers)

    hovertext = (
        df_cities["city"]
        + "<br>Lat: " + df_cities["lat"].round(4).astype(str)
        + "<br>Lon: " + df_cities["lon"].round(4).astype(str)
    )

    markers = go.Scatter3d(
        x=x,
        y=y,
        z=z,
        mode="markers+text",
        text=df_cities["city"],
        textposition="top center",
        hovertext=hovertext,
        hoverinfo="text",
        marker=dict(size=6, color="crimson", line=dict(width=1, color="white")),
    )
    return markers


def main():
    # 1) Earth sphere
    radius = 1.0
    earth = build_earth_sphere(radius=radius, n=120)

    # 2) Cities (London, New York, Tokyo, Sydney)
    cities = pd.DataFrame(
        {
            "city": ["London", "New York", "Tokyo", "Sydney"],
            "lat": [51.5074, 40.7128, 35.6762, -33.8688],
            "lon": [-0.1278, -74.0060, 139.6503, 151.2093],
        }
    )

    # 3) and 4) Convert to 3D and plot markers
    city_markers = build_city_markers(cities, radius=radius, offset=0.015)

    # 5) Build interactive figure
    fig = go.Figure(data=[earth, city_markers])
    fig.update_layout(
        title="Interactive 3D Earth with City Markers",
        scene=dict(
            xaxis=dict(visible=False),
            yaxis=dict(visible=False),
            zaxis=dict(visible=False),
            aspectmode="data",
        ),
        margin=dict(l=0, r=0, t=40, b=0),
    )

    # Show interactive plot (rotatable and zoomable)
    fig.show()
    # Optionally save to HTML
    # fig.write_html("earth_cities.html", auto_open=True)

if __name__ == "__main__":
    main()
```
